Define a funciton that will evaluate the derivative of cos(x) at an arbitrary position x

```python
def dcosx(x,dx,scheme='central'):
    '''Takes the derivative of cos at a position x with a resolution dx either using forward or central difference scheme'''
    from numpy import cos
    if scheme == 'forward':
        return (cos(x+dx)-cos(x))/dx
    if scheme == 'central':
        return (cos(x+dx/2)-cos(x-dx/2))/dx
```

1) Evaluate the derivative at position x using the forward
difference scheme. Plot log(Error)=log10|f’(numerically)-
f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at
x=π/4. You should find Error~dxn; what is n? what is the
optimal value for dx.

The optimal value for dx is 10<sup>-8</sup>.

```python
import numpy as np
x = np.pi/4
error = np.empty(17)
dx = np.empty(17)
fanalytic = -np.sin(x)
for i in range(17):
    dx[i] = 10**(-16+i)
    error[i] = np.absolute(dcosx(x,dx[i],'forward')-fanalytic)
    
logdx = np.log10(dx)
logerror = np.log10(error)
```

```python
import matplotlib.pyplot as plt

plt.plot(logdx,logerror)
plt.ylabel('log(error)')
plt.xlabel('log(dx)')
plt.title('forward difference at x = pi/4')
plt.show()
```
